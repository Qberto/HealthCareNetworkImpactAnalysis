### Project Plan

1. "Feature Engineering"
    
    ○ Patient 
        - Closest doctor - ODCM
        - 2nd closest doctor - ODCM
        - Income value - BA Demographics
        - Elderly count
        - Frequency of visits (Elderly count * index) - Pandas
        - (Refined) - CPSD - Capacity Percent at 2nd closest doctor
        - (Refined) - Services needed
    
    ○ Doctor
        - Amount of patients - ODCM
        - Average drive time - ODCM + Pandas
        - Capacity - Dummy
        - Capacity Percent - Amount of Patients / Capacity
        - Amount of at-risk patients
        - (Refined) - Services provided

2. Produce NLIP score for each patient:
    ○ NLIP: Network Loss Impact to Patient
    ○ NLIP = (Dij to 2nd closest doctor * index) + (Frequency of Visits * index) + (Income Val * index) + (CPSD * Index)

3. Compute total NLIP score for each doctor:
    a. Provider NLIP = Sum. of NLPI scores for patients within Service Area

4. Normalize NLIP scores across state providers

5. Classify normalized NLDI scores into Severe, High, Moderate, Low, Safe values

#### Calculate elderly-derived frequency of visits

```python
def translate_range_function(old_val, old_min, old_max, new_min, new_max):
    new_val = (((old_val - old_min) * (new_max - new_min)) / (old_max - old_min)) + new_min
    return new_val

def translate_range_in_df(dataframe, old_column, new_column_name, new_min, new_max):
    old_min = dataframe[old_column].min()
    old_max = dataframe[old_column].max()
    dataframe[new_column_name] = dataframe.apply(lambda x: translate_range_function(x[old_column], 
                                                                                    old_min, old_max, 
                                                                                    new_min, new_max), axis=1)
    return dataframe
```

```python
out_df = translate_range_in_df(out_patients_final_df, "MEDAGE_CY", "visit_freq", 0, 25)
```

### Calculate NLIP Score
Produce NLIP score for each patient:

NLIP: Network Loss Impact to Patient Score

NLIP = (Dij to 2nd closest doctor * index) + (Frequency of Visits * index) + (Income Val * index) + (CPSD * Index)

```python
# def calculate_nlip(r1_to_r2_imp_delta, ix_r1_to_r2_imp_delta,
#                    visit_freq, ix_visit_freq,
#                    income_val, ix_income_val,
#                    cpsd, ix_cpsd):
#     nlip_val = (r1_to_r2_imp_delta * ix_r1_to_r2_imp_delta) + (visit_freq * ix_visit_freq) + (income_val * ix_income_val) + (cpsd *ix_cpsd)
#     return nlip_val
```

```python
list_of_fields_to_normalize = [
    'R1to2_Impedance_Delta',
    'visit_freq',
    'MEDHINC_CY'
]
```

```python
out_df['R1to2_Impedance_Delta'] = out_df.R1to2_Impedance_Delta.astype(float)
```

```python
for field in list_of_fields_to_normalize:
    print(field)
    print(out_df[field].dtype)
    out_df[field+"_zscore"] = (out_df[field] - out_df[field].mean())/out_df[field].std(ddof=0)
```

```python
def calculate_nlip(r1_to_r2_imp_delta, ix_r1_to_r2_imp_delta,
                   visit_freq, ix_visit_freq,
                   income_val, ix_income_val):
    nlip_val = float((float(r1_to_r2_imp_delta) * ix_r1_to_r2_imp_delta) + (float(visit_freq) * ix_visit_freq) + (float(income_val) * ix_income_val))
    return nlip_val
```

```python
out_df['nlip_val'] = out_df.apply(lambda x: calculate_nlip(x['R1to2_Impedance_Delta'], ix_dij_delta,
                                                             x['visit_freq_zscore'], ix_freq_visits,
                                                             x['MEDHINC_CY_zscore'], ix_income_val),
                                    axis=1)
```

```python
out_df['nlip_zscore'] = (out_df['nlip_val'] - out_df['nlip_val'].mean())/out_df['nlip_val'].std(ddof=0)
```
